```python
#CREATING A FUNCTION TO TRANSFORM TEXT (example 3.17B) INTO NUMBER (REAL NUMBERS, example 3,170,000,000).
d = {
     'k':  3,
     'M':  6,
     'B':  9,
     'T': 12,
     '%': -2
}
def text_to_num(text):
    try:
        if text[-1] in d:
            num, magnitude = text[:-1], text[-1]
            return float(num) * 10 ** d[magnitude]  #this case is when "text" has T, B, M, k or %
        else:
            return float(text) #this case is when "text" is string but look like a numeric
    except:
        try:
            return 1.0*text #this is when "text" is already numeric
        except:
            return 0.0 #it will reach this case when it is impossible to transform into numeric
```

```python
df_FTSEItaliaAllShare['Volume'] = df_FTSEItaliaAllShare['Volume'].apply(text_to_num)
for col in ['Open', 'High', 'Low', 'Close', 'Adj Close']:
    df_FTSEItaliaAllShare[col] = pd.to_numeric(df_FTSEItaliaAllShare[col], errors='coerce')
```

```python
import pandas as pd     

def results_summary_to_dataframe(results, item,reference, date_range=""):
    '''This takes the result of an statsmodel results table and transforms it into a dataframe'''
    prefix = reference + "_"# .replace("-","_").replace("=","_")+"_"
    
#    if(reference == "GC=F"):
#        prefix="g_"
#    else:
#        prefix="btc_"
    pvals = results.pvalues
    coeff = results.params
    conf_lower = results.conf_int()[0]
    conf_higher = results.conf_int()[1]
    
    df_symbols.loc[df_symbols['symbol']==item,prefix+"_daterange"] = date_range    
    df_symbols.loc[df_symbols['symbol']==item,prefix+'c'] = coeff[0]
    df_symbols.loc[df_symbols['symbol']==item,prefix+'p'] = pvals[0]
    df_symbols.loc[df_symbols['symbol']==item,prefix+'c0'] = coeff[1]
    df_symbols.loc[df_symbols['symbol']==item,prefix+'p0'] = pvals[1]
    df_symbols.loc[df_symbols['symbol']==item,prefix+'c1'] = coeff[2]
    df_symbols.loc[df_symbols['symbol']==item,prefix+'p1'] = pvals[2]
    df_symbols.loc[df_symbols['symbol']==item,prefix+'c2'] = coeff[3]
    df_symbols.loc[df_symbols['symbol']==item,prefix+'p2'] = pvals[3]
    df_symbols.loc[df_symbols['symbol']==item,prefix+'c3'] = coeff[4]
    df_symbols.loc[df_symbols['symbol']==item,prefix+'p3'] = pvals[4]

    if(pvals[2]>0.1 and pvals[3]>0.1 and pvals[4]>0.1): 
        """ If one of the
parameters c1, c2 or c3 is significantly different from zero, there is evidence
of a non-linear relationship between gold (Bitcoin) and the G7
stock market. """
       
        df_symbols.loc[df_symbols['symbol']==item,prefix+'conclusion'] = 'neither'#'neither haven nor hedge'
        df_symbols.loc[df_symbols['symbol']==item,prefix+'numeric'] = 0        
    elif(coeff[1]>0.00 and pvals[1] <=0.10) :
        df_symbols.loc[df_symbols['symbol']==item,prefix+'conclusion'] = 'diversi' #'is a diversifier'
        df_symbols.loc[df_symbols['symbol']==item,prefix+'numeric'] = 1        
    elif (coeff[2]<0 and pvals[2] <=0.10) and (coeff[3]<0 and pvals[3] <=0.10) and (coeff[4]<0 and pvals[4] <=0.10):
        df_symbols.loc[df_symbols['symbol']==item,prefix+'conclusion'] = 'haven+' #'is a strong safe haven'
        df_symbols.loc[df_symbols['symbol']==item,prefix+'numeric'] = 2
    elif (coeff[2]<0 or pvals[2] > 0.10) and (coeff[3]<0 or pvals[3] > 0.10) and (coeff[4]<0 or pvals[4] > 0.10):
        df_symbols.loc[df_symbols['symbol']==item,prefix+'conclusion'] = 'haven-' #'is a strong safe haven'
        df_symbols.loc[df_symbols['symbol']==item,prefix+'numeric'] = 3
    elif(coeff[1]<0 and pvals[1]<=0.10) :
        df_symbols.loc[df_symbols['symbol']==item,prefix+'conclusion'] = 'hedge+' #'is a strong hedge'
        df_symbols.loc[df_symbols['symbol']==item,prefix+'numeric'] = 4
    elif(pvals[1]>0.10) :
        df_symbols.loc[df_symbols['symbol']==item,prefix+'conclusion'] = 'hedge-' #'is a weak hedge'
        df_symbols.loc[df_symbols['symbol']==item,prefix+'numeric'] = 5
    else:
        df_symbols.loc[df_symbols['symbol']==item,prefix+'conclusion'] = 'nothing'#'neither haven nor hedge'
        df_symbols.loc[df_symbols['symbol']==item,prefix+'numeric'] = 6        
        
    results_df = pd.DataFrame({"pvals":pvals,
                               "coeff":coeff,
                               "conf_lower":conf_lower,
                               "conf_higher":conf_higher
                                })

    #Reordering...
    results_df = results_df[["coeff","pvals","conf_lower","conf_higher"]]
    return results_df
```

```python
def color_positive_green(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: green'` for positive
    strings, black otherwise.
    """
    try:
        if ('hedge' in val):
            background_color = 'background-color: green;'
            font_style = 'font-style: italic;'
            font_color = 'color:black;'
        elif ('haven' in val):
            background_color = 'background-color: red;'
            font_style = 'font-style: italic;'
            font_color = 'color:black;'
        elif ('diver' in val):
            background_color = 'background-color: orange;'
            font_style = 'font-style: italic;'
            font_color = 'color:black;'
        elif ('neither' in val):
            background_color = 'background-color: white;'
            font_style = 'font-style: normal;'       
            font_color = 'color:black;'
        else:
            background_color = 'background-color: blue;'
            font_style = 'font-style: normal;'
            font_color = 'color:white;'
    except:
        background_color = 'background-color: blue;'
        font_style = 'font-style: normal;'
        font_color = 'color:white;'
        
    return ('{} {} {} text-align: center;').format(background_color,font_style, font_color)
 
df_results_merged.sort_values(by=['Volume (in Trillions of Dollars)','type','Technology',], ascending = [False,True, False]).style.applymap(color_positive_green)
```
